```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='income', y='avg_spend_per_visit', hue='continue_buying')
plt.xlabel('Income')
plt.ylabel('Average Spend Per Visit')
plt.title('Income vs. Average Spend Per Visit by Customer Retention')
plt.legend(title='Continue Buying', loc='upper right')
plt.show()
```

```python

# Convert 'continue_buying' to numerical values for correlation analysis
df['continue_buying_numeric'] = df['continue_buying'].apply(lambda x: 1 if x == 'Yes' else 0)

# Selecting relevant columns for correlation analysis
ratings_columns = ['quality_rating', 'price_rating', 'ambiance_rating', 'service_rating', 'continue_buying_numeric']
correlation_matrix = df[ratings_columns].corr()

# Plotting the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Matrix of Ratings and Continue Buying")
plt.show()
```

```python
# Split 'promo_source' by ';' and create separate rows for each source
df_expanded = df.assign(promo_source=df['promo_source'].str.split(';')).explode('promo_source')
df_expanded['promo_source'] = df_expanded['promo_source'].str.strip()  # Remove any extra whitespace
```

```python
# Calculate retention rate by promo source
retention_by_source = df_expanded.groupby('promo_source')['continue_buying'].value_counts(normalize=True).unstack().fillna(0)
retention_by_source['Retention Rate'] = retention_by_source['Yes']

# Sort by retention rate in descending order
retention_by_source = retention_by_source[['Retention Rate']].sort_values(by='Retention Rate', ascending=False)
print(retention_by_source)
```

```python
# Plotting retention rate by promo source
plt.figure(figsize=(12, 6))
sns.barplot(x=retention_by_source.index, y=retention_by_source['Retention Rate'], palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Promo Source')
plt.ylabel('Retention Rate')
plt.title('Customer Retention Rate by Promo Source')
plt.show()
```

## Observations:

1. **Membership and Repeat Buying:** Customers with memberships are much more likely to keep buying than those without memberships. This shows that having a membership encourages people to stay loyal.

2. **Factors Influencing Retention:** Higher ratings for quality, price, and ambiance are linked to customers sticking around, with quality having the biggest impact. Service ratings also matter, but they donâ€™t have as strong an effect on customer retention.
3. **Customer Retention Rate by Promo Source:** "Biillbords,Emails" have more impact on customer retaintion "Starbucks Website/App ,Deal Sites,Social" Media have less impact as compare to Biillbords,Emails and "Application offer & Never hear" have no impact on Customer retaintion.

## Solutions:

1. **Make Memberships More Appealing:** Adding more value to memberships (like exclusive discounts or rewards) could help keep members coming back and attract new members.

2. **Focus on Quality and Price:** Keeping product quality high and prices competitive can improve customer satisfaction and loyalty, as these are closely tied to repeat buying.

3. **Improve Ambiance:** Enhancing the atmosphere of the space (like dÃ©cor, lighting, and comfort) can also make a difference in keeping customers coming back.

4. **Promo Source:** Spend money only on "Biillbords,Emails,Starbucks Website/App ,Deal Sites,Social Media" except "Application Offer & Never Hear"  for promotions.


## Conclusion:

**The findings show that having a membership greatly increases the chances of customers coming back. People who rate the quality, price, and ambiance highly are also more likely to continue buying, with quality being the most important factor. To keep customers loyal, it would be smart to focus on improving membership perks, keeping quality high, and making sure prices feel fair. Service is still important, but it seems to have less impact on whether customers come back compared to quality, price, and ambiance.**
